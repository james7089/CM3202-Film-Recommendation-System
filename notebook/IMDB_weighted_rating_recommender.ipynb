```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ast import literal_eval
import warnings; warnings.simplefilter('ignore')
```

```python
movies_df = pd.read_csv("TheMoviesDataset\movies_metadata.csv")
```

```python
#reformates genres from [{'id': 1, 'name': 'x'}, {'id': 2, 'name': 'y'}] to [x, y]
movies_df['genres'] = movies_df['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

IMDB's weighted rating formula 

v = number of votes for the movie

m = minimum votes required (e.g. 95th percentile)

R = Average rating of a movie (scale: 1-10)

C = mean vote of all movies

Weighted Rating (WR) = (v Ã· (v+m)) Ã— R + (m Ã· (v+m)) Ã— C

```python
# this is V
vote_counts = movies_df[movies_df['vote_count'].notnull()]['vote_count'].astype('int')

# this is R
vote_averages = movies_df[movies_df['vote_average'].notnull()]['vote_average'].astype('int')

# this is C
C = vote_averages.mean()
```

```python
m = vote_counts.quantile(0.95)
```

```python
def weighted_rating(x):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
movies_df['wr'] = movies_df.apply(weighted_rating, axis=1)
```

```python
#movies ordered by IMDB weighted rating
movies_by_wr = movies_df.sort_values(by='wr', ascending=False)
```
